```python
sns.set_style("darkgrid")

plt.figure(figsize = (12,12))
plt.xlabel('Area in Square Kilometers', fontsize = 15)
plt.ylabel('Name of Installation', fontsize = 15)
plt.title('The Sunny Twenty, Twenty Largest Solar Power Plants by Area, Nation, and Output(in MW)', fontsize = 20)
sns.scatterplot(data=df, x = 'Sq Km' , y = 'Name' , size = 'MWp (Megawatts Peak Capacity)' , hue = 'Country', Alpha = 0.75,  sizes=(50, 1550));
```
